```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import make_blobs
# used to generate datasets doesnot makes clusters
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
# distance based algorightms me standard scaling karnin padti hai neto kuch kafi
# dur banjayenge
```

```python
X,y_true=make_blobs(n_samples=500,centers=4,cluster_std=0.60,random_state=42)
df=pd.DataFrame(X)
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
df=pd.DataFrame(X_scaled)
df.head()
```

```python
wcss=[]
k_range=range(1,11)
```

```python
for k in k_range :
  kmeans=KMeans(n_clusters=k)
  kmeans.fit(X_scaled)
  wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
sns.lineplot(wcss)
```

```python
kmeans_final=KMeans(n_clusters=4)
kmeans_final.fit(X_scaled)
```

```python
y_pred=kmeans_final.predict(X_scaled)
y_pred
```

```python
sns.scatterplot(x=df[0],y=df[1],hue=y_pred)
```

but what does hue=y_pred do ?
In sns.scatterplot(x=df[0], y=df[1], hue=y_pred):

hue=y_pred tells Seaborn to color the data points based on the values in the y_pred array.
Since y_pred contains the cluster labels (0, 1, 2, 3) assigned by the K-Means algorithm, this effectively colors each cluster with a different color.
