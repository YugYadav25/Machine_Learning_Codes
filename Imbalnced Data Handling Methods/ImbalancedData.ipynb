```python
df=pd.read_csv("/content/drive/MyDrive/Machine Learning/ImbalancedDataSet/creditcard.csv")
df['Class'].value_counts() # we can see the data is highly imbalanced towared 1
```

```python
X=df.drop('Class',axis=1)
y=df['Class']
```

```python
df
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,f1_score
from sklearn.model_selection import KFold,GridSearchCV
import numpy as np
```

```python
log_class=LogisticRegression()
grid={'C':10.0**np.arange(-2,3),'penalty':['l1','l2']}
# 'C':10.0**np.arange(-2,3)' gives value from 10 to power -2 to 3

cv=KFold(n_splits=5)
```

```python
grid_search_cv=GridSearchCV(
    estimator=log_class,
    param_grid=grid,
    n_jobs=-1,# uses all the cores of the parameter
    scoring='f1_micro'# scoring parameter
)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
grid_search_cv.fit(X_train,y_train)
```

```python
y_pred=grid_search_cv.predict(X_test)
confusion_matrix(y_test,y_pred)
```

```python
print(accuracy_score(y_test,y_pred),'-->not because your model is performing well but the data is completely imbalanced')
```

```python
f1_score(y_test,y_pred)
```

```python
df_class_0=df[df['Class']==0]
df_class_1=df[df['Class']==1]
```

```python
print(df_class_0.shape)
print(df_class_1.shape)
```

# UnderSampling

```python
df_class_0_sampled=df_class_0.sample(df_class_1.shape[0])
df_class_0_sampled.shape
```

```python
print(df_class_0_sampled.shape)
print(df_class_1.shape )
```

```python
df_balanced=pd.concat([df_class_0_sampled,df_class_1])
df_balanced['Class'].value_counts()
```

# Oversampling

```python
df_class_1_oversampled=df_class_1.sample(df_class_0.shape[0],replace=True)
df_class_1_oversampled.shape
```

```python
df_test_oversampled=pd.concat([df_class_0,df_class_1_oversampled])
df_test_oversampled['Class'].value_counts()
```

# SMOTE METHOD
uses k neart algo

```python
from imblearn.over_sampling import SMOTE
```

```python
df['Class'].value_counts()
```

```python
smote=SMOTE(sampling_strategy='minority')
```

```python
X_train.shape
```

```python
y_train.value_counts()
```

```python
X_sm,y_sm=smote.fit_resample(X_train,y_train)
```

```python
y_sm.value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_sm,y_sm,test_size=0.33,random_state=42,stratify=y_sm)
```

```python
y_pred=grid_search_cv.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
f1_score(y_test,y_pred)
# f1 sc043 improved from 68 to 80
```

```python
linear_model=LogisticRegression()
linear_model.fit(X_train,y_train)
```

```python
y_pred1=linear_model.predict(X_test)
accuracy_score(y_test,y_pred1)
```

```python
f1_score(y_test,y_pred1)
```
