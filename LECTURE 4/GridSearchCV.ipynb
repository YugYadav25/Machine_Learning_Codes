```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('iris')
```

```python
df
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop('species',axis=1)
y=df['species']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model_knn=KNeighborsClassifier(n_neighbors=13)
model_knn.fit(X_train,y_train)
```

```python
model_knn.score(X_test,y_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn import svm
```

```python
model_svm=svm.SVC()
```

```python
classifier=GridSearchCV((model_svm),{
    'C':[1,10,20,30], # comes into the svm model
    'kernel':['rbf','linear']
},cv=5,return_train_score=False) # cv - Determines the cross-validation splitting strategy for the k-fold (splitting
# the test data and train data for better analysis )
```

```python
classifier.fit(X_train,y_train)
```

```python
classifier.cv_results_
```

```python
results = pd.DataFrame(classifier.cv_results_)
results[['param_C','param_kernel','mean_test_score']]
```

```python
classifier2 = GridSearchCV(model_knn, {
    'n_neighbors':[5,9,13],
    'algorithm':['auto', 'ball_tree', 'kd_tree', 'brute']
}, cv=5, return_train_score=False)
```

```python
classifier2.fit(X_train,y_train)
```

```python
from sklearn.ensemble import StackingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import f1_score,accuracy_score,confusion_matrix
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le =LabelEncoder()
base_models=[
    ('dt',DecisionTreeClassifier(random_state=42)),
    ('svc',svm.SVC(probability = True , kernel = 'rbf', random_state=42)),
    ('knn',KNeighborsClassifier(n_neighbors=5))
]
```

```python
  meta_learner=svm.SVC(C=30,random_state=42)
```

```python
stacking_models=StackingClassifier(
     estimators= base_models,
    final_estimator= meta_learner,
    cv= 10
)
```

```python
stacking_models.fit(X_train,y_train)
```

```python
y_pred=stacking_models.predict(X_test)
accuracy=accuracy_score(y_pred,y_test)
```

1. Random Initialization & Shuffling
Machine learning models start with random weights, and data is often shuffled randomly before training. Every run can give slightly different results because of this randomness.
2. train_test_split randomness
If your tutor didn't set a random_state, the data split into training/testing sets will be different every time. Your model trains on different data than theirs.
