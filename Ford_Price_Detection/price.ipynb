```python
X=df.drop(columns=['price'])
y=df['price']
```

```python
X
```

```python
# One Hot Encoding
X_encoded1=pd.get_dummies(X,columns=['model','transmission','fuelType'],drop_first=True)
```

```python
X_encoded1=X_encoded1.astype(int)
```

```python
X_encoded1.head()
```

```python
from sklearn.preprocessing import LabelEncoder

columns = ['model', 'transmission', 'fuelType']

Xlable_Encoded = X.copy()  # make a safe copy
label_encoders = {}

for col in columns:
    le = LabelEncoder()
    Xlable_Encoded[col] = le.fit_transform(Xlable_Encoded[col].astype(str))  # Convert to string in case of nulls
    label_encoders[col] = le
```

```python
Xlable_Encoded
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
numerical_cols = ['year', 'mileage', 'tax', 'mpg', 'engineSize']
scaler =StandardScaler()
```

```python

X_encoded1[numerical_cols]=scaler.fit_transform(X_encoded1[numerical_cols])
X_encoded1
```

```python
Xlable_Encoded[['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg',
       'engineSize']] = scaler.fit_transform(Xlable_Encoded[['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg',
       'engineSize']])
```

```python
Xlable_Encoded

```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
r2=r2_score(y_test,y_pred)
r2
```

```python
n=X_test.shape[0]
p=X_test.shape[1]
adjusted_r2=1-((1-r2)*(n-1))/(n-p-1)
adjusted_r2
```
